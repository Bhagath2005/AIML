```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as bye
import seaborn as hi
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score


titanic = pd.read_csv("titanic_train.csv")
```

```python
print("Data types of each feature:")
print(titanic.dtypes)
```

```python
print("Summary of dataset:")
print(titanic.describe(percentiles=[.25, .50, .75, .90]))
```

```python
print("\n Names of the columns/features in the dataset:")
print(titanic.columns)
```

```python
print("Relative size of survived/un-survived passengers")
hi.countplot(x="Survived", data=titanic)
bye.title("Survival Count")
bye.show()
```

```python
hi.countplot(x="Survived", hue="Sex", data=titanic)
bye.title("Survival Count by Gender")
bye.show()
```

```python
hi.countplot(x="Pclass", hue="Survived", data=titanic)
bye.title("Survival Count by Passenger Class")
bye.show()
```

```python
titanic["Age"].fillna(titanic["Age"].median(), inplace=True)





titanic["Sex"] = titanic["Sex"].map({"male": 0, "female": 1})
titanic["Embarked"].fillna('S', inplace=True)
titanic["Embarked"] = titanic["Embarked"].map({"S": 0, "C": 1, "Q": 2})

print("\nData after wrangling:")
print(titanic.head())
```
